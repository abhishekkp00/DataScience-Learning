### Reference Counting in Python

```python
import sys

a = []

print(sys.getrefcount(a))  # Output: 2 (one for 'a' and one for the argument passed to getrefcount)
```

### Garbage Collection

Python includes a cyclic garbage collector to handle the reference cycle. Reference cycles occurs when objects reference to each other, preventing their references counts from reaching zero

```python
import gc
##enable automatic garbage collection
gc.enable()
```

```python
gc.disable()
```

```python
#manually trigger garbage collection
gc.collect()
```

```python
### get garbage collection stats
print(gc.get_stats())
```

```python
### get the unreachable objects
print(gc.garbage)
```
